# Employees Salaries Exploratory Data Analysis (EDA)

This project analyzes employee salary data to identify salary trends, job title patterns, and compensation insights using Python.

```python
import pandas as pd
import numpy as np
pd.set_option('Display.max_columns',None)
```

```python
df = pd.read_csv('Salaries.csv')
df
```

## Display top 10 rows of the Dataset

```python
# Task 1: Access the top 10 rows of the Dataset
# To access the top 10 Rows of dataset we use head() function
# other alternatives --> df.loc[],df.iloc[] and df[rowstart:rowend:step]
print('Top 10 Rows of the Dataset')
print('----------------------------')
df.head(10)
```

## Check last 10 Rows of dataset

```python
# Task 2: Access last 10 rows in the dataset
# To access last 10 rows we use tail() function
# other ways df.loc,df.iloc and df[len(df)-10]
print('Last 10 rows of the dataset')
print('--------------------------------')
df[len(df)-10:]
```

## Find number of rows and columns in the dataset

```python
# Task 3: Find the number of rows and columns in the dataset
# To find number of rows and columns we use shape attribute
print('Total Number of rows and columns in the dataset')
print('-------------------------------------------------')
for rows,columns in [df.shape]:
    print('Number of rows =',rows)
    print('Number of columns =',columns)
print('-------------------------------------------------')
```

## Get the information about dataset like total number of rows, total number of columns, datatypes of each columns and memory requirement

```python
# Task 4: Find the number of rows/columns and data type of each column along with memory requirement
# To access the memory requirement,number rows/columns, non-null values we use a function info()
df.info()
```

## Check null values in the Dataset

```python
# Task 5: Display the number of null values in each column
# To get the number of null values in each column we can use isna() and isnull() functions
df.isnull().sum()
```

## Drop ID,Notes,Agency and status columns

```python
# Task 6: Drop ID,Notes,Agency and status columns
# To remove the columns or rows we use drop() function 
# Hare axis=1 refers to the columns(axis=0 is for rows)
df = df.drop(['Id','Notes','Agency','Status'],axis=1)
df
```

## Get overall statistics in the dataframe

```python
# Task 7: Get overall statistics in the dataframe
# To get the overall description of the data we use describe() function
df.describe(include='all')
```

## Find the occurrence of the top 5 employee names

```python
# Task 8: Find the occurences of first 5 employee names
# To obtain the number of repeat values in employeename we use head() and value_counts() functions
df['EmployeeName'].head().value_counts()
```

## Find the number of unique job titles 

```python
# Task 9: Find the number of unique job tittles
# To know the number of unique value we use nunique() function 
# nunique() will give number of unique values
print('Number of unique job tittles =',df['JobTitle'].nunique())
```

## Find total number of job titles contain captain

```python
# Task 10: Find total number of job tittles contain captain
# To obtain number of job titles containing captain 
# We should obtain the jobtitles with captain gives the series object
# lenth of the series will give the number of jobtitle with captain
captain = len(df[df['JobTitle'].str.contains('Captain')]['JobTitle'])
print('Number of titles containing Captain =',captain)
```

## Display all the employee names from fire department

```python
# Task 11: Display all the employee names from fire department
# To obtain EmoployeeName from Firedepartment 
# Obtain the rows with of jobtitle with fire department then access the employeenames
print('Employees belongs to Fire Department')
print('----------------------------------------------------------')
df[df['JobTitle'].str.contains('FIRE DEPARTMENT')]['EmployeeName']
```

## Find minimum,maximum,average Base pay

```python
# Task 12: Find minimum,maximum,average Base pay
# To know the min, mean and max values of BasePay
# We use agg() function 
print('Minimum,Maximum and Average values of BasePay')
print('--------------------------------------------------')
df['BasePay'].agg(['min','mean','max'])
    
```

##  Data Cleaning

## Replace 'not provided' in employee name to NaN

```python
# Task 13: Replace 'not provided' in employee name to NaN
# To replace 'not provided' with NaN we use replace() function
# To access those values use isna() or isnull()
df['EmployeeName'] = df['EmployeeName'].replace('Not provided',np.nan)
df[df['EmployeeName'].isna()]
```

## Drop the rows having 5 missing values

```python
# Task 14: Drop the rows having 5 missing values
# To drop the rows with 5 missing values first find the rows with 5 missing values
# Use drop() function to remove rows
ind = df[df.isnull().sum(axis=1) == 5].index
df = df.drop(ind)
df
```

## Find the job title of Albert Pardini

```python
# Task 15: Find the job tittle of Albert Pardini
# To know the job title of Albert Pardini using EmployeeName we can access the jobtitle column
# This gives series output then access the values in the series

print('Job title of Alber Pardini is',df[df['EmployeeName']=='Albert Pardini']['JobTitle'].values)
```

## How much did Albert Pardini earn(Include benefits)

```python
# Task 16: How much Albert Pardini make(Include benefits)
# To find the income of Albert Pardinin including benefits we should access the total benefits column
# this gives a series object using 'values' attribute you can get the TotalPayBenefits
print('Total income of Albert Pardini including benefits =',df[df['EmployeeName']=='Albert Pardini']['TotalPayBenefits'].values)
```

## Display name of the person having highest base pay

```python
# Task 17: Display name of the person having highest base pay
# To get the EmployeeName with highest BasePay we use max()
# Find the maximum of BasePay then access name of the Employee
print('Person with highest BasePay =',df[df['BasePay']==df['BasePay'].max()]['EmployeeName'].values)
```

## Average base pay of all employees per year

```python
# Task 18: Average base pay of all employees per year
# To obtain the average BasePay per year we should use groupby() function for Year column
# Use mean() function to obtain
print('Average BasePay per year')
print('----------------------------')
df2 = df.groupby('Year')
df2['BasePay'].mean()
```

## Average base pay of employee per job title

```python
# Task 19: Average base pay of employees per job title
# To obtain the average BasePay per year we should use groupby() function for jobtitle column
# Use mean() function to obtain
df3 = df.groupby('JobTitle')
df3['BasePay'].mean()
```

## Average base pay of employee having job title Accountant

```python
# Task 20: Average base pay of employees having job title Accountant
# To obtain the BasePay of the employee with jobtitle Accountant
# Access the employees with jobtitle accountant then access their BasePay and use mean() function to obtain average
df[df['JobTitle'].str.contains('Accountant')]['BasePay'].mean()
```

## Find top 5 most common jobs

```python
# Task 21: Find top 5 most common jobs
# To obtain the top 5 most repeated job titles
# Use the value_counts() function and head() function for JobTitle 
df['JobTitle'].value_counts().head()
```
